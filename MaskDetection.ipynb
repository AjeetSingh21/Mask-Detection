```python
import tensorflow as tf
from keras_preprocessing.image import ImageDataGenerator
```

```python
train_prop = ImageDataGenerator(
    rescale=1/255,
    shear_range=0.3,
    zoom_range=0.3,
    rotation_range=30,
    width_shift_range=0.4,
    height_shift_range=0.4,
    horizontal_flip=True,
)
```

```python
train_set = train_prop.flow_from_directory(
    r'D:\ML DATASETS CNN AND ANN\FaceMask\facemask_detection\training_set',
    batch_size=32,
    target_size=(64,64),
    class_mode='binary'
)
```

```python
test_prop = ImageDataGenerator(
    rescale=1/255
)
```

```python
test_set = test_prop.flow_from_directory(
    r'D:\ML DATASETS CNN AND ANN\FaceMask\facemask_detection\test_set',
    target_size=(64,64),
    batch_size=32,
    class_mode='binary'
)
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(64,64,3)))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu'))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dropout(0.3))
```

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(train_set,validation_data=test_set,epochs=25)
```

```python
cnn.summary()
```
